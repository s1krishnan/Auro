```python
ride = pd.read_csv('ride.csv',header= None)
```

```python
ride.head()
```

```python
ride.columns = ['id','seat_id','time_board','time_offboard','onboard_lat','onboard_lon','offboard_lat','offboard_lon','star_rating','rider_first_name','rider_last_name','rider_email','survey_ride_reason','survey_feedback']
ride = ride[['id','time_board','time_offboard','onboard_lat','onboard_lon','offboard_lat','offboard_lon']]
```

```python
ride['time_board'] = pd.to_datetime(ride['time_board'])
ride['time_offboard'] = pd.to_datetime(ride['time_offboard'])
```

```python
ride['date_board'] = ride['time_board'].dt.date
ride['time_board'] = ride['time_board'].dt.time
ride['date_offboard'] = ride['time_offboard'].dt.date
ride['time_offboard'] = ride['time_offboard'].dt.time
```

```python
ride = ride.dropna(subset=['onboard_lat'], how='any')
ride = ride.dropna(subset=['onboard_lon'], how='any')
ride = ride.dropna(subset=['offboard_lat'], how='any')
ride = ride.dropna(subset=['offboard_lon'], how='any')
```
